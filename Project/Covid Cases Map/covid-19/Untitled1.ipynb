```python
import pandas as pd

# Change the path to your CSV file
csv_path = 'us-counties-2020.csv'

# Load the CSV file into a pandas DataFrame object
df = pd.read_csv(csv_path)
```

```python
import os
import json

# Get the absolute path to the current working directory
current_dir = os.path.abspath(os.getcwd())

# Construct the path to the JSON file relative to the current working directory
json_path = os.path.join(current_dir,   'template.json')

# Load the JSON file
with open(json_path) as f:
    data = json.load(f)
```

```python
start_date = '2020-03-01'
end_date = '2020-03-31'
filtered_df = df[(df['date'] >= start_date) & (df['date'] <= end_date)]
```

```python
#drop all rows with NaN values
filtered_df = filtered_df.dropna()

#convert 'rebounds' column from float to integer
filtered_df['fips'] = filtered_df['fips'].astype(int) 
```
