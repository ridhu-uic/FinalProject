```python
import os
import json

# Get the absolute path to the current working directory
current_dir = os.path.abspath(os.getcwd())

# Construct the path to the JSON file relative to the current working directory
json_path = os.path.join(current_dir, '..',  'counties.json')

# Load the JSON file
with open(json_path) as f:
    data = json.load(f)
```

```python
county_list=[]
for county in data["objects"]["counties"]["geometries"]:
    county_list.append(county["id"])
    
```

```python
# Construct the path to the JSON file relative to the current working directory
json_path = os.path.join(current_dir, '..',  'he.json')

# Load the JSON file
with open(json_path) as f:
    data_he = json.load(f)
    
```
