```python
import pandas as pd
import datetime as dt

# set up directory
directory = "us-counties-2021.csv"

f = 'weekly_county2county_2021*.csv'
print(f)
csv_files = glob.glob(f)
```

```python
len(csv_files)
```

```python
# create empty dataframe to store covid cases data
covid_cases = pd.read_csv(directory)
```

```python
covid_cases
```

```python
# loop through files
prev = 0
week = []
data = []
for template_dict in template:
    
    fips_value = template_dict['fips']
    covid_cases = covid_cases[covid_cases["fips"] == fips_value]
    for filename in csv_files:
        if filename.endswith(".csv"):
            # extract date information from file name
            print(filename)
            date_string = filename.split("_")[2:]
            ds = date_string[-1].split(".")
            date_string[-1]=ds[0]
            print(date_string)
            file_date = dt.datetime.strptime("_".join(date_string), "%Y_%m_%d")


            # calculate time interval
            end_date = file_date
            start_date =  end_date + dt.timedelta(days=-6)
            date_format = "%Y-%m-%d"
            week.append(end_date)


            # filter covid cases dataframe and sum cases
            interval_cases = covid_cases[(covid_cases["date"] >= str(start_date)) & (covid_cases["date"] < str(end_date))]
            interval_sum = interval_cases["cases"].max()

            if np.isnan(interval_sum):
                no_of_cases = 0
                data.append(no_of_cases)

            else :
                print("sum :",interval_sum)
                print("prev :",prev)

                no_of_cases = interval_sum - prev
                prev = interval_sum
                data.append(no_of_cases)
                # print results
                print(f"Cases between {start_date.date()} and {end_date.date()}: {no_of_cases}")
    break
            
            
        
```

```python
plot_dict = { "Week" : week,
             "Data" : data   
}
df = pd.DataFrame(plot_dict)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert the week column to a datetime object
df['Week'] = pd.to_datetime(df['Week'], format='%Y-%m-%d')

# Set the Week column as the index
df.set_index('Week', inplace=True)

# Plot the data as a line chart
plt.plot(df.index, df['Data'])

# Set the plot title and axis labels
plt.title('Weekly Data')
plt.xlabel('Week')
plt.ylabel('Data')

# Show the plot
plt.show()
```
